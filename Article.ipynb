```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

ff = "Roboto"
fw = "medium"
fs = 11
plt.rcParams["font.family"] = ff
plt.rcParams["font.weight"] = fw
plt.rcParams["axes.titleweight"] = fw
plt.rcParams["font.size"] = fs
plt.rcParams["axes.titlesize"] = plt.rcParams["font.size"]
plt.rcParams["lines.linewidth"] = 2.5
plt.rcParams["mathtext.fontset"] = "custom"
plt.rcParams['mathtext.default'] = 'it'
plt.rcParams["mathtext.rm"] = ff
plt.rcParams["mathtext.it"] = ff + ":italic"
plt.rcParams["mathtext.bf"] = ff + ":italic"
```

```python
def cm2inch(x,y): return [x/2.54, y/2.54]
```

```python
df = pd.read_csv("./Data.csv")
df
```

```python
fig, ax = plt.subplots(1, 1, figsize=cm2inch(7,5.5), dpi=150)

ax.plot(df["x"], df["y1"], "ko-", ms=7, mfc="w", lw=0, label="$y_1$", zorder=3)
ax.plot(df["x"], df["y1"], "k-", label="$y_1$")
ax.plot(df["x"], df["y2"], "ks-", mfc="w", lw=1, label="$y_2$")
ax.plot(df["x"], df["y3"], "kx-", lw=1, mew=2, label="$y_3$")

ax.set_xlabel("x label", fontweight=fw, labelpad=0)
ax.set_ylabel("y label", fontweight=fw, labelpad=0)
# ax.set_xscale("log")
# ax.set_yscale("log")
# ax.set_xticks([-1, -0.5, 0, 1])
# ax.set_xticklabels(["-1", "-0.5", "0", "1"])
# ax.set_yticks([-1, -0.5, 0, 1])
# ax.set_yticklabels(["-1", "-0.5", "0", "1"])
ax.set_title("Title")
# ax.legend(frameon=False, prop={"size":10}, handlelength=1, labelspacing=0.3
#           bbox_to_anchor=(1.05,1.05), loc="upper right")

fig.tight_layout()
plt.savefig("article-1.pdf", fonttype=42)
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=cm2inch(14,12.5), dpi=150)

for i in range(2):
    for j in range(2):
        ax[i][j].plot(df["x"], df["y1"], "ko-", ms=7, mfc="w", lw=0, label="$y_1$", zorder=3)
        ax[i][j].plot(df["x"], df["y1"], "k-", label="$y_1$")
        ax[i][j].plot(df["x"], df["y2"], "ks-", mfc="w", lw=1, label="$y_2$")
        ax[i][j].plot(df["x"], df["y3"], "kx-", lw=1, mew=2, label="$y_3$")

        ax[i][j].set_xlabel("x label", fontweight=fw, labelpad=0)
        ax[i][j].set_ylabel("y label", fontweight=fw, labelpad=0)
        # ax[i][j].set_xscale("log")
        # ax[i][j].set_yscale("log")
        # ax[i][j].set_xticks([-1, -0.5, 0, 1])
        # ax[i][j].set_xticklabels(["-1", "-0.5", "0", "1"])
        # ax[i][j].set_yticks([-1, -0.5, 0, 1])
        # ax[i][j].set_yticklabels(["-1", "-0.5", "0", "1"])
        ax[i][j].set_title("Title")
        # ax[i][j].legend(frameon=False, prop={"size":10}, handlelength=1, labelspacing=0.3
        #           bbox_to_anchor=(1.05,1.05), loc="upper right")

fig.legend(["sin+rand", "sin+rand", "cos+rand", "rand"], loc=8, ncol=4, fontsize=fs, frameon=False)
fig.tight_layout()
plt.subplots_adjust(bottom=0.16)
plt.savefig("article-2.pdf", fonttype=42)
plt.show()
```
